```python
def pascal_triangle(n,*,step=1):
    if n == 1: 
        return [[1]]
    triangle = [[1],[1, 1]]
    row = [1,1] 
    print(triangle[0])
    if step==1:
        print(triangle[1])
    for i in range(2, n):
        row = [1] + [sum(column) for column in zip(row[1:], row)] + [1]
        triangle.append(row)
        if i%step==0: 
            print(triangle[i])
pascal_triangle(10)
pascal_triangle(10, step = 3)
pascal_triangle(10, 3)
```

```python
import math
def my_sin(x):
    n=10
    s=0
    m=math.radians(x)
    for i in range(n):
        s=s+(-1)**i*m**(2*i+1)/math.factorial(2*i+1)
    return s
    
    

def my_cos(x):
    n=10
    c=0
    m=math.radians(x)
    for i in range(n):
        c=c+(-1)**i*m**(2*i)/math.factorial(2*i)
    return c


def my_exp(x):
    n=35
    e=0
    for i in range(n):
        e=e+x**i/math.factorial(i)
    return e

def my_ln(x):
    t=0
    while x>2:
        x/=my_exp(1)
        t=t+1
    n=10
    l=0
    for i in range(1,n):
        l=l-(-1)**i*(x-1)**i/i
    return l+t
    
print(abs(my_sin(60)-math.sin(math.radians(60))))
print(abs(my_cos(45)-math.cos(math.radians(45))))
print(abs(my_exp(10)-math.exp(10)))
print(abs(my_ln(11)-math.log(11)))
```
